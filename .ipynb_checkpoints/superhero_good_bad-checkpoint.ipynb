# Working with the dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("SuperheroDataset.csv")
```

```python
print(df.columns)
```

```python
truncated_data1 = df[df.columns[4:10]]
truncated_data2 = df[df.columns[16:18]]
truncated_data3 = df[df.columns[21:23]]
```

```python
truncated_data1.reset_index(drop=True, inplace=True)
truncated_data2.reset_index(drop=True, inplace=True)
truncated_data3.reset_index(drop=True, inplace=True)
```

```python
all_df = pd.concat([truncated_data1, truncated_data2, truncated_data3], axis=1)
```

```python
all_df.dropna()#drops all rows with NaN
```

```python
Y = all_df[all_df.columns[6:7]]
```
